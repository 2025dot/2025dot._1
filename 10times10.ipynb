```python
def make_array(color_array):
    img = np.zeros((10,10,3)).astype("uint8")
    for i in range(len(color_array)):
        for j in range(len(color_array[0])):
            if color_array[i][j] == 0:
                img[i,j] = [200,200,200]
            elif color_array[i][j] == 1:
                img[i,j] = [170,220,255]
            elif color_array[i][j] == 2:
                img[i,j] = [170,220,100]
            elif color_array[i][j] == 3:
                img[i,j] = [34,139,34]
            elif color_array[i][j] == 4:
                img[i,j] = [255,255,255]
            elif color_array[i][j] == 5:
                img[i,j] = [255,245,170]
            else:
                img[i,j] = [153,102,51]
    return img
```

```python
# 共通処理にまとめる
def show_with_grid(ax, img, cmap=None):
    ax.imshow(img, cmap=cmap)
    # 軸範囲からサイズ取得
    rows, cols = img.shape[:2]
    ax.set_xticks(np.arange(-0.5, cols, 1))
    ax.set_yticks(np.arange(-0.5, rows, 1))
    ax.grid(which="major", color="black", linestyle="-", linewidth=0.5)
    # 目盛りを消したいなら
    ax.tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)
```
